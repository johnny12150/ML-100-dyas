```python
import pandas as pd
import io
import requests

url = "https://raw.githubusercontent.com/johnny12150/Data_Analytics_Course/master/Lecture%202/Data_and_Images/Batting.csv"
s=requests.get(url).content
PlayerDF = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
PlayerDF.sort_values(by='yearID', axis=0, ascending=True).head(5)
```

```python
PlayerDF.reindex(PlayerDF['yearID']).head(5)
```

```python
PlayerDF['yearID'].isnull().sum()
```

```python
PlayerDF = PlayerDF.dropna(axis=0)
```

```python
PlayerDF['yearID'].isnull().sum()
```

```python
# rename index
PlayerDF_reindex = PlayerDF
PlayerDF_reindex.index = PlayerDF['yearID']
PlayerDF_reindex.head(5)

# same as 
# PlayerDF_reindex = PlayerDF.set_index(PlayerDF['yearID'])
```
