```python
import pandas as pd
import numpy as np
import io
import requests

url = "https://raw.githubusercontent.com/johnny12150/Data_Analytics_Course/master/Lecture%202/Data_and_Images/Batting.csv"
s=requests.get(url).content
PlayerDF = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
PlayerDF.head(5)
```

### 處理類別資料

#### Step 1 Label Encode

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
teamID_one_hot = le.fit_transform(PlayerDF['teamID'])
teamID_one_hot
```

```python
# one-hot後的長度 (相當於資料總數)
len(teamID_one_hot)
```

```python
# 計算nd array的unique數量
count_unique = np.bincount(teamID_one_hot)
count_unique
```

```python
ii = np.nonzero(count_unique)[0]
ii
```

```python
# method 01 印出各個值出現的次數
list(zip(ii,count_unique[ii]))
```

```python
# method 02 印出各個值出現的次數
np.vstack((ii,count_unique[ii])).T
```

```python
# 把轉完的label加到DF
PlayerDF['team_label'] = teamID_one_hot
PlayerDF.head(5)
```

#### Step2 One Hot Encode

```python
one_hot = OneHotEncoder()
teamID_feature_array = one_hot.fit_transform(PlayerDF[['team_label']]).toarray()
feature_labels = list(le.classes_)
list(le.classes_)
```

```python
team_features = pd.DataFrame(teamID_feature_array, columns=feature_labels)
```

```python
team_features.head(5)
```

```python
NewPlayerDF = pd.concat([PlayerDF, team_features], axis=1)
NewPlayerDF.head(5)
```
